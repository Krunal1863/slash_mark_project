```python
import numpy as np
import pandas as pd
from tensorflow.keras.datasets import mnist

# Load MNIST data
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Flatten images to 1D array (28x28 -> 784)
train_images_flattened = train_images.reshape((train_images.shape[0], 28 * 28))
test_images_flattened = test_images.reshape((test_images.shape[0], 28 * 28))

# Convert to DataFrame
train_df = pd.DataFrame(train_images_flattened)
train_df['Label'] = train_labels

test_df = pd.DataFrame(test_images_flattened)
test_df['Label'] = test_labels

# Print the first few rows of the training DataFrame
print(train_df.head())

# Optional: Save as CSV
train_df.to_csv('mnist_train.csv', index=False)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

# Load MNIST data
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Show the Red, Green, and Blue channels of the first image (in full matrix form)
for i in range(1):  # Only for the first image (index 0)
    # Create empty arrays for the R, G, and B channels (28x28 matrix for one image)
    red_channel = np.zeros((28, 28))
    green_channel = np.zeros((28, 28))
    blue_channel = np.zeros((28, 28))
    
    # Assign the grayscale image values to the corresponding channels
    red_channel = train_images[i]  # Red channel
    green_channel = train_images[i]  # Green channel
    blue_channel = train_images[i]  # Blue channel
    
    # Display the matrix for each channel (full matrix without any ellipses)
    print(f"Red Channel (Image {i+1}):")
    print(red_channel)
    
    print(f"\nGreen Channel (Image {i+1}):")
    print(green_channel)
    
    print(f"\nBlue Channel (Image {i+1}):")
    print(blue_channel)
    
    # Visualize the channels
    plt.figure(figsize=(12, 4))

    # Red channel image
    plt.subplot(1, 3, 1)
    plt.imshow(red_channel, cmap='Reds')  # Use 'Reds' colormap for red channel
    plt.title(f'Red Channel Image {i+1}')
    plt.axis('off')

    # Green channel image
    plt.subplot(1, 3, 2)
    plt.imshow(green_channel, cmap='Greens')  # Use 'Greens' colormap for green channel
    plt.title(f'Green Channel Image {i+1}')
    plt.axis('off')

    # Blue channel image
    plt.subplot(1, 3, 3)
    plt.imshow(blue_channel, cmap='Blues')  # Use 'Blues' colormap for blue channel
    plt.title(f'Blue Channel Image {i+1}')
    plt.axis('off')

    # Show all three images
    plt.show()
```
